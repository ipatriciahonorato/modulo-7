### 2. Carregamento de arquivos de dados parquet para dataframes

```python
import os
import pandas as pd

def read_parquet_files(folder_path):
    parquet_files = [file for file in os.listdir(folder_path) if file.endswith('.parquet')]
    
    if not parquet_files:
        print("No Parquet files found in the specified folder.")
        return None
    
    dataframes = []
    common_columns = {}
    
    for file in parquet_files:
        file_path = os.path.join(folder_path, file)
        df = pd.read_parquet(file_path)
        print("File: {} | Shape: {}".format(file, df.shape))
        dataframes.append(df)
        
        for col in df.columns:
            if col not in common_columns:
                common_columns[col] = []
            common_columns[col].append(df[col])
    
    combined_columns = {}
    for col, col_data_list in common_columns.items():
        combined_columns[col] = pd.concat(col_data_list, ignore_index=True)
    
    combined_df = pd.DataFrame(combined_columns)
    
    return combined_df

folder_name = "airplane_data"
flights_combined_data = read_parquet_files(folder_name)
```

### 6. Redução de dimensionalidade das series temporais com codificação de dados de float64 para int64

```python
def convert_columns(df_filtered):
    for coluna in df_filtered.columns:
        if df_filtered[coluna].dtype == 'float64':
            df_filtered[coluna] = df_filtered[coluna].astype('float32')

        if df_filtered[coluna].dtype == 'int64':
            df_filtered[coluna] = df_filtered[coluna].astype('int32')

    return df_filtered
```

```python
flights_combined_data = convert_columns(flights_combined_data)
```
