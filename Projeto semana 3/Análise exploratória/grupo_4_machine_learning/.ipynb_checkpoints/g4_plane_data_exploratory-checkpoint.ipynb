# Pré processamento dos dados

## Porcentagem de colunas com valores NaN

```python
missing_percentage = df_06120089.isnull().sum() / len(df_06120089) * 100
```

```python
threshold = 70 # Linha de corte para números NaN, indica que toda coluna com valores acima disso
```

```python
columns_to_consider = missing_percentage[missing_percentage > threshold].index
```

```python
print("Colunas com mais de {}% de valores NaN:".format(threshold))
for column in columns_to_consider:
    print("Coluna:", column, "| Porcentagem de NaN:", missing_percentage[column], "%")
```

```python
print(df_06120089[columns_to_consider])
```

## Limpeza de dados

### Remoção de linhas com valores nulos

```python
#Cálculo do número de valores nulos por linha
num_nulls_per_row = df_06120089.isnull().sum(axis=1)
```

```python
#Cálculo do número de valores não-nulos por linha:
num_non_nulls_per_row = df_06120089.count(axis=1)
```

```python
#Mascará booleana para comparar os dois números e manter apenas as linhas que possuem mais valores não-nulos que nulos
mask = num_non_nulls_per_row > num_nulls_per_row
```

```python
#Df filtrado com os valores
df_filtered = df_06120089[mask]
```

```python
# Número de linhas antes do filtro
num_of_rows_before = df_06120089.shape[0]
print("Número de linhas antes do filtro:", num_of_rows_before)
```

```python
num_of_rows = df_filtered.shape[0]
print("Número de linhas restantes:", num_of_rows)
```

### Normalização dos dados

```python
# Lista das colunas que você quer verificar
columns_to_check = ['recording_time', 'dateDay-1', 'dateMonth-1', 'dateYear-1', 'phaseOfFlight-1',
                 'message0418DAA-1','message0422DAA-1','amscHprsovDrivF-1a', 'amscHprsovDrivF-1b',
                 'amscHprsovDrivF-2b', 'amscPrsovDrivF-1a',
                 'amscPrsovDrivF-1b', 'amscPrsovDrivF-2b',
                 'basBleedLowPressF-1a', 'basBleedLowPressF-2b',
                 'basBleedLowTempF-1a', 'basBleedLowTempF-2b',
                 'basBleedOverPressF-1a', 'basBleedOverPressF-2b',
                 'basBleedOverTempF-1a', 'basBleedOverTempF-2b',
                 'bleedFavTmCmd-1a', 'bleedFavTmCmd-1b',
                 'bleedFavTmCmd-2a', 'bleedFavTmCmd-2b', 'bleedFavTmFbk-1a',
                 'bleedFavTmFbk-1b', 'bleedFavTmFbk-2b', 'bleedHprsovCmdStatus-1a',
                 'bleedHprsovCmdStatus-1b', 'bleedHprsovCmdStatus-2a',
                 'bleedHprsovCmdStatus-2b', 'bleedHprsovOpPosStatus-1a',
                 'bleedHprsovOpPosStatus-1b', 'bleedHprsovOpPosStatus-2a',
                 'bleedHprsovOpPosStatus-2b', 'bleedMonPress-1a',
                 'bleedMonPress-1b', 'bleedMonPress-2a', 'bleedMonPress-2b',
                 'bleedOnStatus-1a', 'bleedOnStatus-1b', 'bleedOnStatus-2b',
                 'bleedOverpressCas-2a', 'bleedOverpressCas-2b',
                 'bleedPrecoolDiffPress-1a', 'bleedPrecoolDiffPress-1b',
                 'bleedPrecoolDiffPress-2a', 'bleedPrecoolDiffPress-2b',
                 'bleedPrsovClPosStatus-1a', 'bleedPrsovClPosStatus-2a',
                 'bleedPrsovFbk-1a'
]

# Verificar se os valores de cada coluna estão entre 0 e 1
for col in columns_to_check:
    if df_filtered[col].min() >= 0 and df_filtered[col].max() <= 1:
        print(f'A coluna {col} está normalizada entre 0 e 1.')
    else:
        print(f'A coluna {col} NÃO está normalizada entre 0 e 1.')
```

```python
for col in columns_to_check:
    if abs(df_filtered[col].mean()) < 1e-6 and abs(df_filtered[col].std() - 1) < 1e-6:
        print(f'A coluna {col} está normalizada (média 0 e desvio padrão 1).')
    else:
        print(f'A coluna {col} NÃO está normalizada (média 0 e desvio padrão 1).')
```

```python
for column in df_filtered.columns:
    plt.figure()
    stats.probplot(df_filtered[column], dist="norm", plot=plt)
    plt.title(f"Normal Q-Q Plot: {column}")
    plt.xlabel("Theoretical Quantiles")
    plt.ylabel("Sample Quantiles")
    plt.show()
```
