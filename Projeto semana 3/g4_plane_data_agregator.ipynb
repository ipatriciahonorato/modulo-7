<h2><center> Agregação e seleção de séries de dados de arquivos parquet com dados de telemetria de voos</center></h2>


### 1. Importação de bliotecas

```python
import pandas as pd
import os
```

### 2. Carregamento de arquivos de dados parquet para dataframes

```python
file_path = "primeiro_arquivo_zip/"

parquet_files = []
for file in os.listdir(file_path):
    if file.endswith('.parquet'):
        parquet_files.append(file)

# Loop through the files and read each Parquet file
for file_name in parquet_files:
    flights_combined_data = pd.read_parquet(file_path + file_name)
    print("Loading file: " + file_name)
```

### 4. Seleção das colunas de interesse para a análise

```python
# Selecionar apenas essas colunas dos arquivos parquet
filtered_cols = ['recording_time', 'dateDay-1', 'dateMonth-1', 'dateYear-1', 'phaseOfFlight-1',
                 'message0418DAA-1','message0422DAA-1','amscHprsovDrivF-1a', 'amscHprsovDrivF-1b',
                 'amscHprsovDrivF-2b', 'amscPrsovDrivF-1a',
                 'amscPrsovDrivF-1b', 'amscPrsovDrivF-2b',
                 'basBleedLowPressF-1a', 'basBleedLowPressF-2b',
                 'basBleedLowTempF-1a', 'basBleedLowTempF-2b',
                 'basBleedOverPressF-1a', 'basBleedOverPressF-2b',
                 'basBleedOverTempF-1a', 'basBleedOverTempF-2b',
                 'bleedFavTmCmd-1a', 'bleedFavTmCmd-1b',
                 'bleedFavTmCmd-2a', 'bleedFavTmCmd-2b', 'bleedFavTmFbk-1a',
                 'bleedFavTmFbk-1b', 'bleedFavTmFbk-2b', 'bleedHprsovCmdStatus-1a',
                 'bleedHprsovCmdStatus-1b', 'bleedHprsovCmdStatus-2a',
                 'bleedHprsovCmdStatus-2b', 'bleedHprsovOpPosStatus-1a',
                 'bleedHprsovOpPosStatus-1b', 'bleedHprsovOpPosStatus-2a',
                 'bleedHprsovOpPosStatus-2b', 'bleedMonPress-1a',
                 'bleedMonPress-1b', 'bleedMonPress-2a', 'bleedMonPress-2b',
                 'bleedOnStatus-1a', 'bleedOnStatus-1b', 'bleedOnStatus-2b',
                 'bleedOverpressCas-2a', 'bleedOverpressCas-2b',
                 'bleedPrecoolDiffPress-1a', 'bleedPrecoolDiffPress-1b',
                 'bleedPrecoolDiffPress-2a', 'bleedPrecoolDiffPress-2b',
                 'bleedPrsovClPosStatus-1a', 'bleedPrsovClPosStatus-2a',
                 'bleedPrsovFbk-1a']
```

```python
print(len(filtered_cols))
```

```python
flights_combined_data_curated = []

for df in flights_combined_data:
    columns_exist = True
    for col in filtered_cols:
        if col not in flights_combined_data.columns:
            columns_exist = False
            break
    
    if columns_exist:
        flights_combined_data_curated_df = flights_combined_data[filtered_cols]
        flights_combined_data_curated.append(flights_combined_data_curated_df)
```

```python
# Concatenar todos os DataFrames filtrados em um único Df
flights_combined_data_curated = pd.concat(flights_combined_data_curated, ignore_index=True)
```

### 6. Redução de dimensionalidade das series temporais com codificação de dados de float64 para int64

```python
def convert_columns(df_filtered):
    for coluna in df_filtered.columns:
        if df_filtered[coluna].dtype == 'float64':
            df_filtered[coluna] = df_filtered[coluna].astype('float32')

        if df_filtered[coluna].dtype == 'int64':
            df_filtered[coluna] = df_filtered[coluna].astype('int32')

    return df_filtered
```

```python
flights_combined_data_curated = convert_columns(flights_combined_data_curated)
```
